Let $\boldsymbol A \in \mathbb R^{n\times n}$ be a symmetric positive definite matrix and

(c) (**bonus! up to 4 points**) Find $\nabla^2 f(\boldsymbol x)$. Also show that $\nabla^2 f(\boldsymbol x_{\min})$ is (semi)positive definite and  $\nabla^2 f(\boldsymbol x_{\max})$ is (semi)negative definite, where

    \boldsymbol {Ax}_{\min} = \lambda_{\min} \boldsymbol x_{\min}, \quad \boldsymbol {Ax}_{\max} = \lambda_{\max} \boldsymbol x_{\max}.
$$

### YOUR SOLUTION HERE
$\begin{align*}\boldsymbol{(a)} \space \nabla f(\boldsymbol x) = \frac{\cfrac{\partial (\boldsymbol x^\top \boldsymbol{Ax})}{\partial \boldsymbol x} \times \boldsymbol x^\top \boldsymbol x - \boldsymbol x^\top \boldsymbol{Ax} \times \cfrac{\partial (\boldsymbol x^\top \boldsymbol x)}{\partial \boldsymbol x}}{(\boldsymbol x^\top \boldsymbol x)^{2}}\end{align*}$

$\begin{align*}\frac{\partial (\boldsymbol x^\top \boldsymbol A \boldsymbol x)}{\partial \boldsymbol x} = 2 \boldsymbol x^\top \boldsymbol A \end{align*}$

$\begin{align*} \nabla f(\boldsymbol x) = \frac{2 \boldsymbol x^\top \boldsymbol A \times \boldsymbol x^\top \boldsymbol x - \boldsymbol x^\top \boldsymbol{Ax} \times 2 \boldsymbol x^\top}{(\boldsymbol x^\top \boldsymbol x)^{2}} = \frac{\boldsymbol x^\top \boldsymbol x \times 2 \boldsymbol x^\top \boldsymbol A - \boldsymbol x^\top \boldsymbol{Ax} \times 2 \boldsymbol x^\top}{(\boldsymbol x^\top \boldsymbol x)^{2}} = 0 \end{align*}$

$\begin{align*}\boldsymbol{Ax} = \boldsymbol{\lambda x} \end{align*}$

$\begin{align*}\nabla f(\boldsymbol x) = \frac{2 \boldsymbol x^\top \boldsymbol A \times \boldsymbol x^\top \boldsymbol x - \boldsymbol x^\top \boldsymbol{Ax} \times 2 \boldsymbol x^\top}{(\boldsymbol x^\top \boldsymbol x)^{2}} = \frac{2 \boldsymbol x^\top \times \boldsymbol x^\top \boldsymbol {Ax} - \boldsymbol x^\top \boldsymbol{Ax} \times 2 \boldsymbol x^\top}{(\boldsymbol x^\top \boldsymbol x)^{2}} = \frac{2 \boldsymbol x^\top \times \boldsymbol x^\top \boldsymbol {\lambda x} - \boldsymbol x^\top \boldsymbol{\lambda x} \times 2 \boldsymbol x^\top}{(\boldsymbol x^\top \boldsymbol x)^{2}} = 0\end{align*}$

$\begin{align*}\boldsymbol{(b)} \space f(\boldsymbol x) = \frac{\boldsymbol x_{i}^\top \boldsymbol{A_{i} x_{i}}}{\boldsymbol x_{i}^\top \boldsymbol x_{i}} = \boldsymbol \lambda_{i} \end{align*}$

```python
import numpy as np
import timeit
import matplotlib.pyplot as plt

%matplotlib inline

# Value for dimensions
n = 1000

# For Numpy example with 3x3 matrices and 3x1 vectors multiplication
A = np.random.randint(5, size=(3, 3))
x = np.random.randint(5, size=(3, 1))

print(f"Example for 3x3 matrix and 3x1 vector: \nMatrix A: \n{A} \nvector x: \n{x}\n\n")

# For pure Python example with 3x3 matrices and 3x1 vectors multiplication
A_list = A.tolist()
x_list = x.tolist()
b_list = []

def purePyMult(A_list, x_list: list, size: int):
    temp = 0

    for i in range(size):
        for j in range(size):
            temp += A_list[i][j] * x_list[j][0]
        b_list.append(temp)
        temp = 0
    return b_list


def numPyMult(A, x: np.ndarray):
    return np.dot(A, x)


# For comparing results of 3x3 matrix and 3x1 vector with Python and Numpy
def printRes():
    b_list = purePyMult(A_list, x_list, 3)
    b = numPyMult(A, x)
    
    print(f"Python result: \n{b_list} \nNumpy result: \n{b}\n")
        

# Getting the lists filled with execution times
def getExecTimes(n):
    
    pyTimeList = []
    npTimeList = []

    for i in range(1, n+1):
        # Increasing the matrix and vector sizes
        A = np.random.randint(100, size=(i, i))
        x = np.random.randint(100, size=(i, 1))
        A_list = A.tolist()
        x_list = x.tolist()

        start = timeit.default_timer()
        b_list = purePyMult(A_list, x_list, i)
        end = timeit.default_timer()
        pyTimeList.append(end-start)

        start = timeit.default_timer()
        b = numPyMult(A, x)
        end = timeit.default_timer()
        npTimeList.append(end-start)

    return pyTimeList, npTimeList

printRes()
pyTimeList, npTimeList = getExecTimes(n)

# Plotting (may take some time, change n in case you are tired of waiting)
plt.plot(np.arange(1, n+1), pyTimeList, c='r', lw=2, label="py")
plt.plot(np.arange(1, n+1), npTimeList, c='b', lw=2, label="np")
plt.xlim(1, n)
plt.xlabel("Size")
plt.ylabel("Time (s)")
plt.legend()
plt.title("Matrix-Vector Multiplication: Python vs NumPy")
plt.grid(ls=":")
plt.show()
```

## Demo for task 1.5: dot product performance

```python
import numpy as np
import matplotlib.pyplot as plt
from time import time

%config InlineBackend.figure_format = 'svg'


def python_inner(a, b: np.array):
    return sum(a * b)

        
def measure_time(func, size, n_samples=1):
    result = np.zeros(n_samples)
    for i in range(n_samples):
        begin = time()
        func(np.random.randn(size), np.random.randn(size))
        result[i] = time() - begin
    return result.mean()
    

def check_correctness(func, max_size, attempts=10):
    for size in range(1, max_size + 1):
        for _ in range(attempts):
            a = np.random.randn(size)
            b = np.random.randn(size)
            assert np.allclose(func(a, b), np.dot(a, b)), f"different values for {a} and {b}"
    

def get_times_lists(func, step=20, max_size=1000, n_samples=20):
    py_times = []
    np_times = []
    sizes = np.arange(20, max_size + 1, 20)
    for size in sizes:
        py_times.append(measure_time(func, size))
        np_times.append(measure_time(np.dot, size))
    return np.array(py_times), np.array(np_times)
```

```python
# TESTING AREA

check_correctness(python_inner, 100)
```

```python
# PLOTTING AREA

def plot_time_vs_size(step=20, max_size=1000, n_samples=20):
    inner_py_times, inner_np_times = get_times_lists(python_inner, step, max_size, n_samples)
    sizes = np.arange(20, max_size + 1, 20)
    plt.plot(sizes, inner_py_times, c='r', lw=2, label="py")
    plt.plot(sizes, inner_np_times, c='b', lw=2, label="np")
    plt.xlim(0, max_size)
    plt.title("Dot product: python vs numpy")
    plt.legend()
    plt.xlabel("size")
    plt.ylabel("time, s")
    plt.grid(ls=":");


plot_time_vs_size()
```

On average, `np.dot` is 2-3 times better, than pythonic loops:

```python
inner_py_times, inner_np_times = get_times_lists(python_inner)
plt.plot(inner_py_times / inner_np_times, lw=2, c='r');
plt.ylim(0, 5)
plt.grid(ls=":")
```
