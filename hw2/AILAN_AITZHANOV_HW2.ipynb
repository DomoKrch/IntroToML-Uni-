## Task 1.1 (1 point)

Let $\xi \sim \mathrm{Bin}(n, p)$. Prove that $\mathbb E\xi = np$, $\mathbb V\xi = np(1-p)$.

### YOUR SOLUTION HERE
$\begin{align*} \boldsymbol{(a)} \space \mathbb E(\xi) = \sum_{k=0}^n kP(\xi=k) =  \sum_{k=1}^n k \dbinom{n}{k} p^{k} q^{n-k}\end{align*}$

$\begin{align*} \sum_{k=1}^n \bcancel{k} \frac{n(n-1)!}{\bcancel{k}(k-1)!(n-k)!} p p^{k-1} q^{(k-1)-(q-1)}\end{align*}$

$\begin{align*} \sum_{k=1}^n np \dbinom{n-1}{k-1} p^{k-1} q^{(n-1)-(k-1)} \end{align*}$

$\begin{align*} i = k-1 \space and \space j = n-1 \end{align*}$

$\begin{align*} np \sum_{i=0}^j \dbinom{j}{i} p^{i} q^{j-i} \end{align*}$

$\begin{align*} \boldsymbol{Binomial \space theorem}: \sum_{i=0}^j \dbinom{j}{i} p^{i} q^{j-i} => (p+q)^{j} = 1^j \end{align*}$

$\begin{align*} np \sum_{i=0}^j \dbinom{j}{i} p^{i} q^{j-i} = np \end{align*}$

$\begin{align*} \mathbb E(\xi) = np \end{align*}$

$\begin{align*}\end{align*}$

$\begin{align*} \boldsymbol{(b)} \space \mathbb V(\xi) = \mathbb E(\xi^{2}) - \mathbb E(\xi)^{2} \end{align*}$

$\begin{align*} \mathbb E(\xi)^{2} = \sum_{k=0}^n k^{2}P(\xi=k) = \sum_{k=1}^n k^{2} \dbinom{n}{k} p^{k} q^{n-k} \end{align*}$

$\begin{align*} \sum_{k=1}^n k^{\bcancel{2}} \frac{n(n-1)!}{\bcancel{k}(k-1)!(n-k)!} p p^{k-1} q^{(n-1)-(k-1)} \end{align*}$

$\begin{align*} np \sum_{k=1}^n k \dbinom{n-1}{k-1} p^{k-1} q^{(n-1)-(k-1)}\end{align*}$

$\begin{align*} i = k - 1 \space and \space j = n - 1 \end{align*}$

$\begin{align*} np \sum_{i=0}^j (i+1) \dbinom{j}{i} p^{i} q^{j-i} \end{align*}$

$\begin{align*} np (\sum_{i=0}^j i \dbinom{j}{i} p^{i} q^{j-i} + \sum_{i=0}^j \dbinom{j}{i} p^{i} q^{j-i}) \end{align*}$

$\begin{align*} np (\sum_{i=1}^j j \dbinom{j-1}{i-1} pp^{i-1} q^{(j-1)-(i-1)} + 1) \end{align*}$

$\begin{align*} np^{2} (n-1) \sum_{i=1}^j \dbinom{j-1}{i-1} p^{i-1} q^{(j-1)-(i-1)} + np \end{align*}$

$\begin{align*} m = i - 1 \space and \space l = j - 1 \end{align*}$

$\begin{align*} np^{2} (n-1) \sum_{m=0}^l \dbinom{l}{m} p^{m} q^{l-m} + np\end{align*}$

$\begin{align*} np^{2} (n-1) + np = (np)^{2} - np^{2} + np \end{align*}$

$\begin{align*} \space \mathbb V(\xi) = \mathbb E(\xi^{2}) - \mathbb E(\xi)^{2} = \bcancel{(np)^{2}} - np^{2} + np - \bcancel{(np)^{2}} = np(1-p)\end{align*}$

## Task 1.2 (2 points)

$$
    p(x) = \frac1{\Gamma(\alpha) \beta^\alpha} x^{\alpha - 1} e^{-\frac x \beta}.
$$

*Hint*. The answer is expressed in terms of **digamma function** $\psi(t) = \frac{\Gamma'(t)}{\Gamma(t)}$.

### YOUR SOLUTION HERE

$\begin{align*} \mathbb H(x) = - \int_{- \infty}^\infty f(x) \log(f(x)) \,dx \space and \space \mathbb E(x) = \int_{- \infty}^\infty f(x) x \,dx => \mathbb H(x) = - \mathbb E(\log_{\beta}(P(x)))\end{align*}$

$\begin{align*} - \mathbb E(\log_{\beta}(P(x))) = - \mathbb E(\log_{\beta}(\frac1{\Gamma(\alpha) \beta^\alpha} x^{\alpha - 1} e^{-\frac x \beta})) \end{align*}$

$\begin{align*} - \mathbb E((\alpha - 1) \log_{\beta}(x) - \frac{x}{\beta} \log_{\beta}(e) - \log_{\beta}(\Gamma(\alpha)) - \alpha \log_{\beta}(\beta)) \end{align*}$

$\begin{align*} (1 - \alpha) \mathbb E(\log_{\beta}(x)) + \frac{\log_{\beta}(e)}{\beta} \mathbb E(x) + \log_{\beta}(\Gamma(\alpha)) + \alpha = (1 - \alpha) \mathbb E(\log_{\beta}(x)) + \frac{\log_{\beta}(e)}{\bcancel{\beta}} \alpha \bcancel{\beta} + \log_{\beta}(\Gamma(\alpha)) + \alpha \end{align*}$

$\begin{align*} \mathbb E(\log_{\beta}(x)) = \psi(\alpha) + \log_{\beta}(\beta) = \psi(\alpha) + 1 \space => \space \psi(\alpha) + 1 - \alpha \psi(\alpha) - \bcancel{\alpha} + \alpha \log_{\beta}(e) + \log_{\beta}(\Gamma(\alpha)) + \bcancel{\alpha} \end{align*}$

$\begin{align*} \mathbb H(x) = \psi(\alpha)(1 - \alpha) + \alpha \log_{\beta}(e) + \log_{\beta}(\Gamma(\alpha)) + 1 \end{align*}$

## Task 1.3 (3 points)

Let $X_1, \ldots, X_n$ is an i.i.d. sample from $\mathcal N(0, \sigma^2)$ and $\widehat \sigma = \frac 1n \sum\limits_{i=1}^n |X_i|$.

* Find bias, se and MSE of the estimation $\widehat \sigma$. Is this estimation unbiased? 
* If $\widehat \sigma$ turned out to be biased, fix it to get an unbiased estimation of $\sigma$. Find se of this corrected estimation. Is it consistent?

### YOUR SOLUTION HERE
$\begin{align*} \boldsymbol{a)} \space bias(\hat{\sigma}) = \mathbb E(\hat{\sigma}) - \sigma  = \mathbb{E} (\frac{1}{n} \sum_{i=1}^n |x_{i}|) - \sigma \end{align*}$

$\begin{align*} \sigma = \sqrt{\frac{1}{n} \sum_{i=1}^n (x_{i} - \mu)^{2}} = \sum_{i=1}^n \frac{|x_{i}|}{\sqrt{n}} \end{align*}$

$\begin{align*} bias(\hat{\sigma}) = \frac{\sigma}{\sqrt{n}} - \sigma = \sigma (\frac{1}{\sqrt{n}} - 1) \ \neq 0 => \boldsymbol{biased}\end{align*}$

$\begin{align*} SE(\hat{\sigma}) = \sqrt{Var(\hat{\sigma})}= \sqrt{Var(\frac{1}{n} \sum_{i=1}^n |x_{i}|)} = \sqrt{\frac{\sigma^{2}}{n}} = \frac{\sigma}{\sqrt{n}} \end{align*}$

$\begin{align*} MSE = (bias(\hat{\sigma}))^{2} + (SE(\hat{\sigma}))^{2} = (\sigma (\frac{1}{\sqrt{n}} - 1))^{2} + (\frac{\sigma}{\sqrt{n}})^{2} = \frac{2 \sigma^{2}}{n} - \frac{2 \sigma}{\sqrt{n}} + \sigma^{2} \end{align*}$

$\begin{align*} \end{align*}$

$\begin{align*} \boldsymbol{b)} \space \mathbb E(\hat{\sigma}) = \sigma \end{align*}$

$\begin{align*} bias(\hat{\sigma}) = \sigma - \sigma = 0 => \boldsymbol{unbiased} \end{align*}$

$\begin{align*} SE(\hat{\sigma}) = \sqrt{Var(\hat{\sigma})} = \sqrt{\frac{1}{n} Var(\sum_{i=1}^n |x_{i}|)} = \sqrt(\frac{1}{\bcancel{n}} * \bcancel{n} \sigma^{2}) = \sigma \end{align*}$

$\begin{align*} MSE = (bias(\hat{\sigma}))^{2} + (SE(\hat{\sigma}))^{2} = 0 + \sigma^{2} = \sigma^{2} \end{align*}$

$\begin{align*} \lim_{n \to \infty} \sigma^{2} = \sigma^{2} => \boldsymbol{inconsistent} \end{align*}$

## Task 1.4 (3 points)

Let $X_1, \ldots, X_n$ is an i.i.d. sample form $U[0, 2\theta]$. The parameter $\theta$ can be estimated as
$\widehat\theta = \overline X_n$ (sample average) or $\tilde\theta = \mathrm{med}(X_1, \ldots, X_n)$ (sample median). According to ML Handbook, both estimations are unbiased, and $\mathbb V \widehat\theta = \frac{\theta^2}{3n}$, $\mathbb V \tilde\theta = \frac{\theta^2}{n+3}$. Hence, due to the central limit theorem

$$
    \widehat\theta \approx \mathcal N\Big(\theta, \frac{\theta^2}{3n}\Big), \quad
    \tilde\theta \approx \mathcal N\Big(\theta, \frac{\theta^2}{n+3}\Big), \quad n \gg 1.
$$

In this exercise you are suggested to verify this equalities numerically using NumPy, and plot the results (see example below).

### Possible result of the experiment

```python
%config InlineBackend.figure_format = 'svg'

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


font = {'family' : 'serif',
        'size'   : 17,
        'weight' : 'normal'
       }

plt.rc('font', **font)

plt.rc('text', usetex=True)
plt.rc('xtick', labelsize=18)
plt.rc('ytick', labelsize=18)
plt.rc('axes', titlesize=20)
plt.rc('legend', fontsize=18)
plt.rc('text.latex', preamble=r'\usepackage[utf8]{inputenc}')
plt.rc('text.latex', preamble=r'\usepackage[T2A]{fontenc}')
plt.rc('text.latex', preamble=r'\usepackage[russian]{babel}')
plt.rc('text.latex', preamble=r'\usepackage{amsmath}')

def plot_hists(theta: float, n, n_samples: int):
    sigma = (theta**2) / (3 * n)
    mu = theta

    # taken from the docs: https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html
    first_theta = np.random.normal(mu, sigma, n_samples)
    count, bins, ignored = plt.hist(first_theta, 30, density=True)
    plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=4, color='r', label="average")

    mu = theta
    sigma = (theta**2) / (n + 3)
    
    second_theta = np.random.normal(mu, sigma, n_samples)
    count, bins, ignored = plt.hist(second_theta, 30, density=True)
    plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=4, color='b', label="median")


    plt.xlabel("Data points")
    plt.ylabel("Density")
    plt.legend()
    plt.title("Sample average vs median")
    plt.grid(ls=":")
    plt.show() 
    

plot_hists(theta=1.5, n=2000, n_samples=4000)
```
